# Basic Implementation of Linear Regression (using first principles)

```python
class LinearRegression:
    def __init__(self):
        pass
    def train_gradient_desc(self, X,y, learning_rate=0.02, n_iter=100, Ïµ = 1e-4):
        #Initialize the params
        n_samples, n_features = X.shape
        self.weights = np.zeros(shape=(n_features,1))
        self.bias= 0
        cost = []
        
        for i in range(n_iter):
            #compute linear combination of the input
            y_predict = np.dot(X, self.weights) + self.bias
            
            # cost over training set
            costs = (1/n_samples)*np.sum((y_predict-y)**2)
            cost.append(costs)
            
            if i%100 ==0:
                print(f"cost at iteration {i}: {costs}")
             #compute gradients   
            Dj_dw = (2/n_samples)*np.dot(X.T, (y_predict - y))
            Dj_db = (2/n_samples)*np.sum((y_predict-y))
            # reg term
            self.weights = self.weights - learning_rate*Dj_dw
            #new_weights = self.weights - learning_rate*Dj_dw
            self.bias = self.bias-learning_rate*Dj_db
            
           # if np.sum(new_weights)
            
        return self.weights, self.bias, cost
    
    
    
    
    def predict(self,X):
        return np.dot(X,self.weights) + self.bias
            
                
        
```

```python
regressor = LinearRegression()
w_trained, b_trained, cost = regressor.train_gradient_desc(X, y, learning_rate=0.005, n_iter=1000)
```

```python
plt.plot(np.arange(1000), cost, label = r"$\alpha = 0.005$")
plt.title("Show of cost func during training")
plt.xlabel("No. of iterations")
plt.ylabel("Cost J(Î¸)")
plt.legend()
plt.show()
```
